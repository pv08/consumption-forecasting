```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gru_MAE_Data = pd.read_csv("methodology/reports/All_Features/ALL_MAE_GRU.csv", sep=',')
RNN_MAE_Data = pd.read_csv("methodology/reports/All_Features/ALL_MAE_RNN.csv", sep=',')
LSTM_MAE_Data = pd.read_csv("methodology/reports/All_Features/ALL_MAE_LSTM.csv", sep=',')
PCA_LSTM_MAE_Data = pd.read_csv("methodology/reports/All_Features/PCA_MAE_LSTM.csv", sep=',')
RNN_Metrics = pd.read_csv("methodology/reports/All_Features/Local/ALL/ALL_Metrics_RNN.csv", sep=';')
TST_Metrics = pd.read_csv("methodology/reports/All_Features/Local/ALL/ALL_Metrics_TST.csv", sep=',')
SHAP_LSTM_Metrics = pd.read_csv("methodology/reports/All_Features/Local/Shap/SHAP_Metrics_LSTM.csv", sep=',')


gru_MSE_Data = pd.read_csv("methodology/reports/All_Features/ALL_MSE_GRU.csv", sep=',')

RNN_MSE_Data = pd.read_csv("methodology/reports/All_Features/ALL_MSE_RNN.csv", sep=',')
LSTM_MSE_Data = pd.read_csv("methodology/reports/All_Features/ALL_MSE_LSTM.csv", sep=',')
PCA_LSTM_MSE_Data = pd.read_csv("methodology/reports/All_Features/PCA_MSE_LSTM.csv", sep=',')
```

```python
RNN_Metrics
```

```python
plt.figure(figsize=(20, 10))
plt.title(f'Baseline Models MAE', fontsize=15)
plt.plot(gru_MAE_Data.index, gru_MAE_Data['participant_id: 661_test_30 - val/rmse'], '^', ls='-',label='GRU')
plt.plot(RNN_Metrics.index, RNN_Metrics['val/val_mae'],'1', ms = 13, ls = '-', c = '#A52A2A',label='RNN')
plt.plot(LSTM_MAE_Data.index, LSTM_MAE_Data['participant_id: 661_test_30 - val/rmse'], ',', ms = 13, ls = '-', c = '#2E8B57', label='LSTM')
plt.xlabel('epochs', fontsize=15)
plt.ylabel('MAE', fontsize=15)
plt.legend(fontsize=15)
plt.savefig(f'methodology/imgs/results/metrics/local_baseline_models_MAE.eps', dpi=600, bbox_inches='tight')
plt.savefig(f'methodology/imgs/results/metrics/local_baseline_models_MAE.svg', dpi=600, bbox_inches='tight')
plt.savefig(f'methodology/imgs/results/metrics/local_baseline_models_MAE.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.title(f'MAE between PCA and normal training', fontsize=15)
plt.plot(LSTM_MAE_Data.index, LSTM_MAE_Data['participant_id: 661_test_30 - val/rmse'], ',', ms = 13, ls = '-', c = '#2E8B57', label='LSTM')
plt.plot(PCA_LSTM_MAE_Data.index, PCA_LSTM_MAE_Data['participant_id: 661_test_30_pca - val/val_mae'],'1', ms = 13, ls = '-', c = '#A52A2A',label='PCA-LSTM')

plt.xlabel('epochs', fontsize=15)
plt.ylabel('MAE', fontsize=15)
plt.legend(fontsize=15)
plt.savefig(f'methodology/imgs/results/metrics/PCA_LSTM_MAE.eps', dpi=600, bbox_inches='tight')
plt.savefig(f'methodology/imgs/results/metrics/PCA_LSTM_MAE.svg', dpi=600, bbox_inches='tight')
plt.savefig(f'methodology/imgs/results/metrics/PCA_LSTM_MAE.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.title(f'MAE between PCA, normal and SHAP training', fontsize=15)
plt.plot(LSTM_MAE_Data.index, LSTM_MAE_Data['participant_id: 661_test_30 - val/rmse'], ',', ms = 13, ls = '-', c = '#2E8B57', label='LSTM')
plt.plot(PCA_LSTM_MAE_Data.index, PCA_LSTM_MAE_Data['participant_id: 661_test_30_pca - val/val_mae'],'1', ms = 13, ls = '-', c = '#A52A2A',label='PCA-LSTM')
plt.plot(SHAP_LSTM_Metrics.index, SHAP_LSTM_Metrics['val/val_mae'],'1', ms = 13, ls = '-', c = '#080FFF',label='SHAP-LSTM')

plt.xlabel('epochs', fontsize=15)
plt.ylabel('MAE', fontsize=15)
plt.legend(fontsize=15)
plt.savefig(f'methodology/imgs/results/metrics/SHAP_LSTM_MAE.eps', dpi=600, bbox_inches='tight')
plt.savefig(f'methodology/imgs/results/metrics/SHAP_LSTM_MAE.svg', dpi=600, bbox_inches='tight')
plt.savefig(f'methodology/imgs/results/metrics/SHAP_LSTM_MAE.png', dpi=600, bbox_inches='tight')
plt.show()
```
