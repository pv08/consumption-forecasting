```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import xgboost as xg
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
from sklearn.preprocessing import MinMaxScaler
from xgboost import XGBRegressor
from src.utils.functions import mkdir_if_not_exists, write_test_json, write_json_file, read_json_file
from typing import Tuple
```

```python
def descale(descaler, values):
    values_2d = np.array(values)[:, np.newaxis]
    return descaler.inverse_transform(values_2d).flatten()
```

```python
df = pd.read_csv('data/Pecanstreet/participants_data/1hour/features/661_test_30_all_features.csv')
df.tail()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
scaler = MinMaxScaler(feature_range=(-1, 1))
scaler = scaler.fit(df)
data = pd.DataFrame(scaler.transform(df), index=df.index, columns=df.columns)
```

```python
scaler.min_[4]
```

```python
X = data.copy()
X.drop('consumption', axis=1, inplace=True)
X.head()
```

```python
y = data['consumption']
y.describe().transpose()
```

```python
n = len(data)
X_train, y_train = X[:int(0.7*n)], y[:int(0.7*n)]
X_validation, y_validation = X[int(.7 * n): int(n * (1.1 - .2))], y[int(.7 * n): int(n * (1.1 - .2))]
X_test, y_test = X[int(n * (1.1 - .2)):], y[int(n * (1.1 - .2)): ]
```

```python
print(X.shape, y.shape)
print(X_train.shape, y_train.shape)
print(X_validation.shape, y_validation.shape)
print(X_test.shape, y_test.shape)
```

```python
param_grid = {
    'max_depth':[3,4,5],
    'learning_rate': [0.1, .01, .05],
    'gamma': [0, .25, 1],
    'reg_lambda': [0, 1, 10],
    'scale_pos_weight': [1, 3, 5]
}
second_param_grid = {
    'max_depth':[4],
    'learning_rate': [0.1, .5, 1],
    'gamma': [.25],
    'reg_lambda': [10, 20, 100],
    'scale_pos_weight': [3]
}


optimal_params = GridSearchCV(estimator=XGBRegressor(seed=42,
                                                          subsample=0.9,
                                                          colsample_bytree=.5),
                             param_grid=param_grid,
                             scoring='neg_mean_squared_error',
                             verbose=0,
                             n_jobs=10,
                             cv=10)

optimal_params.fit(X_train, 
                   y_train, 
                    eval_set=[(X_validation, y_validation)],
                  verbose=False)
print(optimal_params.best_params_)
```

```python
xg_regressor = XGBRegressor(seed=0, objective='reg:squarederror',
                            gamma=1, 
                            learning_rate=0.1, 
                            max_depth=4, 
                            reg_lambda=0, 
                            scale_pos_weight=1)
```

```python
xg_regressor.fit(X_train, y_train, 
                 eval_set=[(X_train, y_train), (X_validation, y_validation)], 
                 eval_metric=['rmse', 'mae', 'mape'],
                
                 verbose=False)
```

```python
y_preds = xg_regressor.predict(X_test)
```

```python
evals_results = xg_regressor.evals_result()
```

```python
train_evals_results = dict(evals_results['validation_0'])
train_evals_results['mse'] = [rmse_error ** 2 for rmse_error in train_evals_results['rmse']]
validation_evals_results = dict(evals_results['validation_1'])
validation_evals_results['mse'] = [rmse_error ** 2 for rmse_error in validation_evals_results['rmse']]
```

```python
plt.title('XGBoost Losses')
plt.plot(train_evals_results['mse'], '-.', label='Train')
plt.plot(validation_evals_results['mse'], label='Validation')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.savefig(f'etc/imgs/participants/Pecanstreet/single-step/661/1hour/XGBoost/train_val_losses.eps', dpi=600, bbox_inches='tight')
plt.savefig(f'etc/imgs/participants/Pecanstreet/single-step/661/1hour/XGBoost/train_val_losses.svg', dpi=600, bbox_inches='tight')
plt.savefig(f'etc/imgs/participants/Pecanstreet/single-step/661/1hour/XGBoost/train_val_losses.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
bst = xg_regressor.get_booster()
for importance_type in ('weight', 'gain', 'cover', 'total_gain', 'total_cover'):
    print('%s: ' % importance_type, bst.get_score(importance_type=importance_type))

node_params = {'share': 'box',
              'style': 'filled, rounded',
              'fillcolor': '#78cbe'}

leaf_params = {'shape': 'box',
              'style': 'filled',
              'fillcolor': '#e48038'}

render = xg.to_graphviz(xg_regressor, num_trees=0, size="10,10",
               condition_node_params=node_params,
               leaf_node_params=leaf_params)
```

```python
render.graph_attr = {'dpi':'600'}
render.render('etc/imgs/participants/Pecanstreet/single-step/661/1hour/XGBoost/xgboost_tree', format = 'svg')
render.render('etc/imgs/participants/Pecanstreet/single-step/661/1hour/XGBoost/xgboost_tree', format = 'eps')
render.render('etc/imgs/participants/Pecanstreet/single-step/661/1hour/XGBoost/xgboost_tree', format = 'png')
render
```

```python
result = [{
    'model': 'XGBoost',
    'MSE': mean_squared_error(y_test, y_preds),
    'MAE': mean_absolute_error(y_test, y_preds),
    'MAPE': mean_absolute_percentage_error(y_test, y_preds)
}]
metrics = pd.DataFrame(result)
metrics
```

```python
def save_json_metrics(content, path, filename, model):
    try:
        data = read_json_file(path, filename)
        models = [result['model'] for result in data]
        if model not in models:
            data.append(content)
            write_json_file(data, path, filename)
            return "[*] - Content inserted"
        else:
            for result in data:
                if result['model'] == model:
                    result = content
                    write_json_file(data, path, filename)
                    return "[*] - Content updated"
    except:
        write_json_file(content, path, filename)
        return "[*] - New Content created"
```

```python
mkdir_if_not_exists('etc/results/Pecanstreet/single-step/661/1hour/XGBoost')
mkdir_if_not_exists('etc/imgs/participants/Pecanstreet/single-step/661/1hour/XGBoost')
save_json_metrics(content=result, 
                  path='etc/results/Pecanstreet/single-step/661/1hour/',
                 filename='metrics_report',
                 model='XGBoost')
```

```python
test_preds = []
for preds, labels in zip(list(y_preds), y_test.to_list()):
    test_preds.append(dict(
        model='XGBoost',
        label=float(labels),
        model_output=float(preds)
    ))
model_infer = pd.DataFrame(test_preds)
model_infer.label = descale(descaler, model_infer.label)
model_infer.model_output = descale(descaler, model_infer.model_output)
model_infer
```

```python
model_infer.to_csv('etc/results/Pecanstreet/single-step/661/1hour/XGBoost/infer.csv')
```
