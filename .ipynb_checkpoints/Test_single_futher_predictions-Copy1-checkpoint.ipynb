```python
import torch as T
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
from src.utils.functions import create_sequences
from torch.utils.data import DataLoader
from src.dataset import PecanDataset
from src.pecan_dataport.participant_preprocessing import PecanParticipantPreProcessing
from src.regressors.lstm_regressor import ConsumptionLSTMRegressor
from sklearn.preprocessing import MinMaxScaler
```

# Função
Função para colocar na medida real do banco

```python
def descale(descaler, values):
    values_2d = np.array(values)[:, np.newaxis]
    return descaler.inverse_transform(values_2d).flatten()
```

# Estado inicial
1 - Pegar o melhor *checkpoint* do modelo
2 - Pegar o scaler que foi usado para os sets de treinamento, validação e teste. No final usar esse scaler para retornar o valor real do conjunto de teste


```python
ckpt = 'etc/ckpts/participants/661_test_30_all/sigmoid/LSTM/best/best-LSTM-chpkt-pecanstreet-participant-id-661_test_30_all_epoch=220.ckpt'
device = T.device('cuda' if T.cuda.is_available() else 'cpu')
```

```python
pecan_dataset = PecanParticipantPreProcessing('661_test_30_all', 'data/participants_data/1min',
                                                   60, task='test')
default_scaler = pecan_dataset.scaler
```

## Forecasting
3 - Trazer os dados que serão previstos fora do conjunto de teste. Esses dados representaram estimativas, onde a feature de consumo será substituída pelo output da previsão da janela anterior

```python
original_data = pd.read_csv("data/participants_data/1min/features/661_features.csv", sep=',')
model_features = pd.read_csv("data/participants_data/1min/features/661_test_30_pca_features.csv", sep=',')
```

## Scaler
4 - Utilizando a íntegra do banco de dados e selecionando a janela até onde prever, será feito um novo scaler. Caso seja utilizado o scaler que foi utilizado nos conjuntos treinamento, validação e teste o valores, quando desnormalizados, serão desproporcionais ao real. Assim sendo, foi selecionado 215 leituras a frente do final do conjunto de teste e foi criado um novo scaler com essas leituras

```python
scaler_df = original_data.iloc[:129301].copy()
```

```python
to_forecast = original_data.iloc[129086:129301].copy()
forecast_scaler = MinMaxScaler(feature_range=(-1, 1))
forecast_scaler = forecast_scaler.fit(scaler_df)
to_forecast = pd.DataFrame(forecast_scaler.transform(to_forecast), index=to_forecast.index,
                         columns=to_forecast.columns
)
```

```python
to_forecast.describe().transpose()
```

```python
#para pegar as última sequencia de previsões
test_df = pecan_dataset.test_df.copy()
initial_predicition = test_df.iloc[-61:].copy()
initial_sequence = create_sequences(initial_predicition, 'consumption', 60)
len(initial_sequence)
```

```python
forecasting_data = to_forecast.copy()
forecasting_confer = forecasting_data.copy()
forecasting_data
```

```python
forecasting_dataset_sequence = PecanDataset(initial_sequence, device)
forecasting_module = DataLoader(dataset=forecasting_dataset_sequence,
                                                 batch_size=1, shuffle=False, num_workers=6, pin_memory=False)
```

```python
model = ConsumptionLSTMRegressor.load_from_checkpoint(checkpoint_path=ckpt,
                                                          strict=False,
                                                          device=device,
                                                          n_features=pecan_dataset.get_n_features(),
                                                          lr=1e-5,
                                                          n_hidden=256,
                                                          n_layers=3,
                                                          dropout=0.3,
                                                          activation_function='sigmoid',
                                                          bidirectional=False, scaler=pecan_dataset.scaler)
model.freeze()
model
```

```python
forecasting_starting_point = 129086
out_limit_pred = []
```

```python
for i in range(215): #prever os próximos 15 minutos
    batch = next(iter(forecasting_module))
    _, output = model(batch['sequence'].to(device))
    out_limit_pred.append(output.item())
    forecasting_data.at[forecasting_starting_point, 'consumption'] = output.item()
    test_df.loc[forecasting_starting_point] = forecasting_data.iloc[i]
    next_prediction = test_df.iloc[-61:].copy()
    next_sequence = create_sequences(next_prediction, 'consumption', 60)

    next_dataset_sequence = PecanDataset(next_sequence, device)

    forecasting_module = DataLoader(dataset=next_dataset_sequence,
                                             batch_size=1, shuffle=False, num_workers=6, pin_memory=False)

    forecasting_starting_point += 1
```

```python
with open(f'etc/results/test/661_test_30_all/result_report.json') as json_file:
    data = json.load(json_file)
    json_file.close()

test_predictions_data = data 
new_prediction_data = []
for model in test_predictions_data:
    for prediction in model['predict']:
        prediction['model'] = model['model']
    new_prediction_data += model['predict']
complete_prediction_df = pd.DataFrame(new_prediction_data)
labels = np.array(complete_prediction_df[complete_prediction_df['model'] == 'GRU'].label)
```

```python
descaler = MinMaxScaler()
descaler.min_, descaler.scale_ = default_scaler.min_[-1], default_scaler.scale_[-1]

descaler_to_foreacast = MinMaxScaler()
descaler_to_foreacast.min_, descaler_to_foreacast.scale_ = forecast_scaler.min_[-1], forecast_scaler.scale_[-1]
```

```python

descaled_lstm_prediction = descale(descaler, complete_prediction_df[complete_prediction_df['model'] == 'LSTM'].model_output)
descaled_label = descale(descaler, labels)
```

```python
descaled_out_limit_pred = descale(descaler_to_foreacast, out_limit_pred)
```

```python
indexes = complete_prediction_df[complete_prediction_df['model'] == 'LSTM'].index[-120:].to_list()
indexes[0], indexes[len(indexes) - 1]
```

```python
out_limit_indexes = np.arange(indexes[len(indexes) - 1], indexes[len(indexes) - 1] + 215, 1)
out_limit_indexes.shape
```

```python
descaled_test = descale(descaler, forecasting_confer['consumption'].values)
descaled_test.shape
```

```python
plt.figure(figsize=(20, 10))
plt.title(f'Relationship between prediction and real data for recorrent models', fontsize=15)
plt.plot(indexes, descaled_label[-120:], '^', ls='-',label='real')
plt.plot(indexes, descaled_lstm_prediction[-120:], '.', ms = 10, ls = '-', c = '#00FFB2',label='LSTM')
plt.plot(out_limit_indexes, descaled_out_limit_pred, '.', ms = 10, ls = '-', c = '#FF0808',label='out-limit')
plt.plot(out_limit_indexes, descaled_test, '.', ms = 10, ls = '-', c = '#1BFF08',label='out-limit-real')



# plt.plot(output, 'm--', label='predictions')

plt.xlabel('time [`min`]', fontsize=15)
plt.ylabel('consumption [`kw`]', fontsize=15)
plt.legend(fontsize=15)
# plt.savefig(f'methodology/imgs/results/forecasting/others_models_forecasting.eps', dpi=600, bbox_inches='tight')
# plt.savefig(f'methodology/imgs/results/forecasting/others_models_forecasting.svg', dpi=600, bbox_inches='tight')
plt.grid()
plt.show()
```
