```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from src.utils.graphs import saveModelMetrics
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, mean_absolute_error, r2_score
```

```python
default_models = ['ConvRNN', 'FCN', 'GRU', 'LSTM', 'MLP', 'ResNet', 'RNN', 'TCN', 'Transformer', 'TST']
default_models = dict(zip(default_models, [None for model in default_models]))
default_models
```

```python
pecan_dict = {
    'all': {
        '1hour': default_models,
        '15min': default_models,
        '1min': default_models
    },
    'PCA': {
        '1hour': default_models,
        '15min': default_models,
        '1min': default_models
    }
}
hue_dict = {
    'all': {
        '1hour': default_models
    },
    'PCA': {
        '1hour': default_models
    }
}
```

```python
for typ in pecan_dict.keys():
    for res in pecan_dict[typ].keys():
        info_data = {}
        for model in default_models:
            data = pd.read_csv(f'etc/results/Pecanstreet/single-step/661/{res}/{typ}/{model}/test_preds.csv')
            info_data.update({'model': model, 'label': data['label'], 'out': data['model_output']})
        metrics = pd.read_json(f'etc/results/Pecanstreet/single-step/661/{res}/{typ}/metrics_report.json')
        info_data.update({'metrics': metrics})
        pecan_dict[typ][res] = info_data
    for res in hue_dict[typ].keys():
        info_data = {}
        for model in default_models:
            data = pd.read_csv(f'etc/results/HUE/single-step/1/{res}/{typ}/{model}/test_preds.csv')
            info_data.update({'model': model, 'label': data['label'], 'out': data['model_output']})
        metrics = pd.read_json(f'etc/results/HUE/single-step/1/{res}/{typ}/metrics_report.json')
        info_data.update({'metrics': metrics})
        hue_dict[typ][res] = info_data
```

```python
all_df = pd.DataFrame({
                'Pecanstreet - 1min': pecan_dict['all']['1min'],
                'Pecanstreet - 15min': pecan_dict['all']['15min'],
                'Pecanstreet - 1hour': pecan_dict['all']['1hour'],
                'HUE - 1hour': hue_dict['all']['1hour'],
})
pca_df = pd.DataFrame({
                'Pecanstreet - 1min': pecan_dict['PCA']['1min'],
                'Pecanstreet - 15min': pecan_dict['PCA']['15min'],
                'Pecanstreet - 1hour': pecan_dict['PCA']['1hour'],
                'HUE - 1hour': hue_dict['PCA']['1hour'],
})
```

```python
f, axs = plt.subplots(nrows=4, ncols=1, figsize=(10, 15))
for i, ax in enumerate(axs.flat):
    all_data = all_df.iloc[:,i]
    pca_data = pca_df.iloc[:,i]
    all_data.dropna(inplace=True)
    pca_data.dropna(inplace=True)
    ax.plot(np.arange(0, 50), all_data['label'][-50:], marker='.', label='Real')
    ax.plot(np.arange(0, 50), all_data['out'][-50:], marker='X', label='Previso',
                  c='#ff7f0e')
    ax.plot(np.arange(0, 50), pca_data['out'][-50:], marker='o', label='Previso - PCA',
                  c='#27AE60')
    ax.set_title(all_df.columns[i], fontweight='bold')
    ax.set_ylabel('Consumo')
    ax.legend()
plt.xlabel('Amostras')
plt.tight_layout()
# plt.savefig(f'etc/imgs/dataset_comparison/SARIMAX_preds.eps', dpi=200, bbox_inches='tight')
# plt.savefig(f'etc/imgs/dataset_comparison/SARIMAX_preds.pdf', dpi=200, bbox_inches='tight')
# plt.savefig(f'etc/imgs/dataset_comparison/SARIMAX_preds.svg', dpi=200, bbox_inches='tight')
# plt.savefig(f'etc/imgs/dataset_comparison/SARIMAX_preds.png', dpi=200, bbox_inches='tight')
plt.show()
```
